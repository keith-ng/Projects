```python
import pandas as pd
```

```python
df = pd.read_csv('eleague_raw.csv')
```

```python
df = df.dropna(axis=1)
```

```python
df
```

```python
### clean match column
df[['team1', 'temp']] = df['match'].str.split(' vs ', 1, expand=True)
df[['team2', 'score']] = df['temp'].str.split('(', 1, expand=True)
df[['t1_score', 't2_score']] = df['score'].str.split('-', 1, expand=True)
df['t2_score'] = df['t2_score'].str.replace(')','')
df = df.drop(columns=['match', 'temp', 'score'])
```

```python
### clean game1 column
df[['map1_pick', 'temp']] = df['game1'].str.split(' pick ', 1, expand=True)
df[['map1', 'map1_score']] = df['temp'].str.split(';', 1, expand=True)
df[['map1_t1', 'map1_t2']] = df['map1_score'].str.split('-', 1, expand=True)
df['map1_t1'] = df['map1_t1'].str.replace('(','')
df['map1_t2'] = df['map1_t2'].str.replace(')','')
df['map1_pick'] = df['map1_pick'].str[3:]
df = df.drop(columns=['game1', 'temp', 'map1_score'])
```

```python
### clean game2 column
df[['map2_pick', 'temp']] = df['game2'].str.split(' pick ', 1, expand=True)
df[['map2', 'map2_score']] = df['temp'].str.split(';', 1, expand=True)
df[['map2_t1', 'map2_t2']] = df['map2_score'].str.split('-', 1, expand=True)
df['map2_t1'] = df['map2_t1'].str.replace('(','')
df['map2_t2'] = df['map2_t2'].str.replace(')','')
df['map2_pick'] = df['map2_pick'].str[3:]
df = df.drop(columns=['game2', 'temp', 'map2_score'])
```

```python
### clean game3 column
df[['map3', 'map3_score']] = df['game3'].str.split('(', 1, expand=True)
df['map3'] = df['map3'].str[3:]
df['map3'] = df['map3'].str.strip(' ;')
df['map3_score'] = df['map3_score'].str.replace(')','')
df[['map3_t1', 'map3_t2']] = df['map3_score'].str.split('-', 1, expand=True)
df = df.drop(columns=['game3', 'map3_score'])
```

```python
df
```

```python
### functions

# retrieve games involving the specified teams
def results(team1_name, team2_name):
    df2 = df[(df.team1 == team1_name) | (df.team2 == team1_name)]
    df3 = df[(df.team1 == team2_name) | (df.team2 == team2_name)]
    df4 = pd.concat([df2,df3])
    return df4
```
